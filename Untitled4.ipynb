### Extracting Specific Sales Records using Indexing and Slicing

We'll now use `pandas`' powerful indexing and slicing capabilities to select specific sales records from our DataFrame `df`. We'll demonstrate a few common scenarios:

1.  **Extracting records for a specific Item Code.**
2.  **Extracting records for a specific Category Name.**
3.  **Slicing to get the first N sales records.**

```python
# Example 1: Extract records for a specific Item Code
# Let's pick the first item code from our `item_code` array
specific_item_code = item_code[0]
item_sales_records = df.loc[df["Item Code"] == specific_item_code]

print(f"Sales records for Item Code {specific_item_code}:")
display(item_sales_records)

# Example 2: Extract records for a specific Category Name
# Let's pick the first category name from our `category_name` array
specific_category_name = category_name[0]
category_sales_records = df.loc[df["Category Name"] == specific_category_name]

print(f"\nSales records for Category Name '{specific_category_name}':")
display(category_sales_records)

# Example 3: Slicing to get the first 5 sales records
first_five_records = df.iloc[:5]

print("\nFirst 5 sales records:")
display(first_five_records)
```
